```python
from seaborn.palettes import mpl_palette
sns.countplot(x='Education', data=market_clean, saturation=5, palette='PuBuGn')
sns.despine(offset=7, trim=True);
plt.title('Education')
```

```python
plt.figure(figsize=(11, 9))
sns.heatmap(market_clean.corr(), annot=True, cmap='PuBuGn',)
plt.title('Correlation Heatmap')
```

```python
df_cmp = market_clean[['Age', 'Income', 'Kidhome', 'Teenhome', 'MntWines', 'NumWebPurchases', 'NumStorePurchases', 'NumWebVisitsMonth', 'NumDealsPurchases']]
correlation_matrix = df_cmp.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='PuBuGn', fmt='.2f', linewidths=0.75)
plt.title('Correlation Heatmap')
```

```python
df_cmp = market_clean[['MntGoldProds', 'MntFishProducts', 'MntFruits', 'MntSweetProducts', 'MntWines', 'MntMeatProducts', 'NumStorePurchases']]
correlation_matrix = df_cmp.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='PuBuGn', fmt='.2f', linewidths=0.75)
plt.title('Correlation Heatmap')
```
